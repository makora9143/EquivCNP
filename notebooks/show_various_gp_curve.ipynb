```python
c1 = '#08AEEA'
c2 = '#2AF598'
c3 = '#FF9000'
```

```python
plt.figure(figsize=(16, 4))
plt.subplot(131)
gp = RBFCurve(train=False, max_total=20, data_range=(-2, 2))
(xc, yc), (xt, yt) = gp.sample(0, 10, 5)
plt.plot(xt, yt, color=c1, label='sample1')
plt.plot(xc, yc, 'o', color=c1)
(xc, yc), (xt, yt) = gp.sample(0, 10, 5)
plt.plot(xt, yt, color=c2, label='sample2')
plt.plot(xc, yc, 'o', color=c2)
(xc, yc), (xt, yt) = gp.sample(0, 10, 5)
plt.plot(xt, yt, color=c3, label='sample3')
plt.plot(xc, yc, 'o', color=c3)
plt.legend()
plt.title("RBF Kernel", fontsize=20)

plt.subplot(132)
gp = MaternCurve(train=False, max_total=20, data_range=(-2, 2))
(xc, yc), (xt, yt) = gp.sample(0, 10, 5)
plt.plot(xt, yt, color=c1, label='sample1')
plt.plot(xc, yc, 'o', color=c1)
(xc, yc), (xt, yt) = gp.sample(0, 10, 5)
plt.plot(xt, yt, color=c2, label='sample2')
plt.plot(xc, yc, 'o', color=c2)
(xc, yc), (xt, yt) = gp.sample(0, 10, 5)
plt.plot(xt, yt, color=c3, label='sample3')
plt.plot(xc, yc, 'o', color=c3)
plt.legend()
plt.title("Matern-5/2 Kernel", fontsize=20)

plt.subplot(133)
gp = PeriodicCurve(train=False, max_total=20, data_range=(-2, 2))
(xc, yc), (xt, yt) = gp.sample(0, 10, 5)
plt.plot(xt, yt, color=c1, label='sample1')
plt.plot(xc, yc, 'o', color=c1)
(xc, yc), (xt, yt) = gp.sample(0, 10, 5)
plt.plot(xt, yt, color=c2, label='sample2')
plt.plot(xc, yc, 'o', color=c2)
(xc, yc), (xt, yt) = gp.sample(0, 10, 5)
plt.plot(xt, yt, color=c3, label='sample3')
plt.plot(xc, yc, 'o', color=c3)
plt.legend()
plt.title("Periodic Kernel", fontsize=20)
```
